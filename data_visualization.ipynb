```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
df = pd.read_csv("dataset/clean_data.csv")

df = df.round(decimals=2)

df.columns
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
fig = sns.distplot(df['cores'], kde=False)
plt.title("Distribution of cores")
plt.show()

fig = sns.distplot(df['input'], kde=False)
plt.title("Distribution of inputs")
plt.show()

fig = sns.distplot(df['average_runtime'], kde=False)
plt.title("Distribution of Avg. Runtime")
plt.show()

fig = sns.distplot(df['speedup'], kde=False)
plt.title("Distribution of speedups")
plt.show()

fig = sns.distplot(df['mem_ram'], kde=False)
plt.title("Distribution of Memory RAM")
plt.show()

fig = sns.distplot(df['efficiency'], kde=False)
plt.title("Distribution of Efficiency")
plt.show()

plt.show()
```

```python
plt.figure(figsize=(10, 7))

sns.heatmap(df.corr(),
            annot = True,
            fmt = '.2f',
            cmap='Blues')
```
