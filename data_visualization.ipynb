```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset/clean_data.csv")
```

```python
NUM_INPUTS = len(df['input'].unique())

plt.figure(figsize=(8,4))
plt.ylabel("cores")
plt.xlabel("Average runtime (s)")

for i in range(NUM_INPUTS):
    df_aux = df[df['input'] == i]    
    plt.plot(df_aux['average_runtime'], df_aux['cores'], label=f'Input {i}')

plt.legend()
plt.show()
```
